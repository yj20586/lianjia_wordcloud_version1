```python
import jieba    #分词包
import numpy    #numpy计算包
import codecs   #codecs提供的open方法来指定打开的文件的语言编码，它会在读取的时候自动转换为内部unicode 
import pandas   
import matplotlib.pyplot as plt
%matplotlib inline

from wordcloud import WordCloud#词云包
```

```python
segment=[]
for item1 in users:
    segs=jieba.cut(item1)
    for item2 in list(segs):
        if len(item2.strip()) > 1:
            segment.append(item2)
```

```python
words_df=pandas.DataFrame({'segment':segment})
words_stat=words_df.groupby(by=['segment'])['segment'].agg({"计数":numpy.size})
words_stat=words_stat.reset_index().sort(columns="计数",ascending=False)
```

```python
wordcloud=WordCloud(font_path="simhei.ttf",background_color="white")
txt_freq = []
for item in words_stat.values:
    txt_freq.append(list(item))
d = {}
for item in txt_freq:
    d[item[0]] = item[1]
wordcloud.fit_words(d)  
plt.imshow(wordcloud)
# frequencies = {u'知乎':100,u'小段同学':4,u'曲小花':3,u'中文分词':2}
# wordcloud = wordcloud.fit_words(frequencies)
# plt.imshow(wordcloud)
```
